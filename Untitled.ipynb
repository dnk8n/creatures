```python
%matplotlib inline
import matplotlib.pyplot as plt


def plot_strategy_counts(data):
    """Plot strategy counts for each iteration."""

    # Extract strategy names
    strategy_names = list(data[0].keys())

    # Create a color map for each strategy
    colors = ["blue", "orange", "green", "red"]
    color_map = dict(zip(strategy_names, colors))

    # Create lists to store data for each strategy
    strategy_counts = {}
    for strategy in strategy_names:
        strategy_counts[strategy] = []

    # Extract counts for each strategy for each iteration
    for iteration in data:
        for strategy in strategy_names:
            strategy_counts[strategy].append(iteration[strategy])

    # Create plot
    plt.figure(figsize=(10, 6))
    for strategy in strategy_names:
        plt.plot(
            range(len(data)),
            strategy_counts[strategy],
            label=strategy,
            color=color_map[strategy],
        )
    plt.xlabel("Iterations")
    plt.ylabel("Strategy Count")
    plt.legend()
    plt.show()
```

```python
plot_strategy_counts(env.state.population_by_strategy)
```

```python
plot_strategy_counts(env.state.hp_by_strategy)
```

```python
plot_strategy_counts(env.state.xp_by_strategy)
```
